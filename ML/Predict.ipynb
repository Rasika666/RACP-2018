```python
import pandas as pd
```

```python
original_training = pd.read_csv('./ad_datasir.csv')
original_training.head()
```

```python
x = original_training.copy()
del x['class']
x.head()
```

```python
y = []
for data in original_training['class'] :
    y.append(data)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
clf.fit(x,y)
```

```python
test_data = pd.read_csv('./ad_data_test.csv')
test_data.head()
```

```python
x_test = test_data.copy()
del x_test['class']
x_test.head()
```

```python
clf.predict(x_test)
```
